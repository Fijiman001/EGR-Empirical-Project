```python
# prompt: I want to load 2 different df

import pandas as pd

# Assuming your dataframes are in CSV files named 'df1.csv' and 'df2.csv'
df1 = pd.read_csv('bond_universe_cleaned_with_wkn.csv')
df2 = pd.read_csv('all_bonds_data_cleaned.csv')

print(df1)
print(df2)
```

```python
# Merge df2 with df1 based on the "WKN" column
merged_df = pd.merge(df1, df2, on='WKN', how='left')

merged_df
```

```python
#remove na in price column
merged_df = merged_df.dropna(subset=['Last Price'])
merged_df
```

```python
# prompt: give me the max of the last Price column

max_last_price = merged_df['Last Price'].max()
max_last_price
merged_df['Last Price'].min()
```
